## Analyze the results
* based on : https://openmm.github.io/openmm-cookbook/latest/notebooks/tutorials/Histone_methyltransferase_simulation_with_multisite_Zn2_structural_ions.html


```python
import mdtraj as md
import matplotlib
matplotlib.use('Agg')
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
# load trajectory
traj = md.load('./results/HSP90_trajectory.dcd', top = './data/hsp90_adp_mg/input.prmtop', stride=1)
# use stride 1 * 1000 (step) * 2 fp(per step) = 2 ps = 0.002  ns
# since we have 0.5 ns, we have 250 frames
traj = traj.atom_slice(traj.top.select('protein or resname ZNB'))
```

```python
# calculate RMSD to first frame and plot figure
rmsd = md.rmsd(traj, traj)
```

```python
plt.figure()
plt.plot(rmsd)
plt.title('RMSD to first frame')
plt.xlabel('Frame (2 ps/frame)')
plt.ylabel('RMSD (nm)')
#plt.savefig('rmsd.png', dpi=300)
#plt.close()
```
